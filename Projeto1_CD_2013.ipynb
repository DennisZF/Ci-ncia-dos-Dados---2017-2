```python
#bibliotecas inicializadas 
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
from numpy import arange
```

```python
#carregando a base de dados
dados2013 = pd.read_table("DOM2013.csv", encoding="windows-1252", sep=",")
```

```python
#tamanho da base de dados
dados2013.shape
```

```python
#criando as categorias
dados2013.V0201 = dados2013.V0201.astype('category')
dados2013.V4105 = dados2013.V4105.astype('category')
dados2013.V2103 = dados2013.V2103.astype('category')
dados2013.V2105 = dados2013.V2105.astype('category')
dados2013.V2107 = dados2013.V2107.astype('category')
dados2013.V2109 = dados2013.V2109.astype('category')
```

```python
#Conversão legenda - Descrição 
dados2013.V0201.cat.categories = ('Particular Permanente','Particular Improvisado','Coletivo')
dados2013.V4105.cat.categories = ('URBANA - Cidade ou vila, área urbanizada','URBANA - Cidade ou vila, área não urbanizada','URBANA - Área urbana isolada','RURAL - Aglomerado rural de extensão urbana','RURAL - Aglomerado rural, isolado, povoado','RURAL - Aglomerado rural, isolado, núcleo','RURAL - Aglomerado rural, isolado, outros aglomerados','RURAL - Zona rural exclusive aglomerado rural')
dados2013.V2103.cat.categories = ('Sim','Não')
dados2013.V2105.cat.categories = ('Sim','Não')
dados2013.V2107.cat.categories = ('Sim','Não')
dados2013.V2109.cat.categories = ('Sim','Não')
```

```python
#1° recorte: estado de são paulo
tabsp= dados2013[dados2013.UF==35]

```

```python
#visualizando todas as colunas disponíveis na base de dados (verificar dicionário para identificação de cada coluna) 
list(tabsp)
```

```python
#corte da base de dados, pegando apenas as colunas desejadas
tabspcortada = tabsp.iloc[:,[6,8,59,72,81,82,83,84]]
```

```python
#retirando valores indefinidos
tabsprenda = tabspcortada[(tabspcortada.V4614 != 999999999999)]
tabsprenda.dropna()
```

```python
#analisando numericamente
#Moradores
contmora = (tabsprenda.V0105.value_counts(sort=True, normalize=True)*100).round(decimals=2)
contmora
```

```python
#Quanto a moradia
contcasa = (tabsprenda.V0201.value_counts(sort=False, normalize=True)*100).round(decimals=2)
contcasa
```

```python
#Quanto a zona
contzona = (tabsprenda.V4105.value_counts(sort=False, normalize=True)*100).round(decimals=2)
contzona
```

```python
#Os moradores tiveram a preocupação de que os alimentos acabassem antes de poderem comprar ou receber mais comida
contprec = (tabsprenda.V2103.value_counts(sort=False, normalize=True)*100).round(decimals=2)
contprec


```

```python
#Os alimentos acabaram antes que os moradores deste domicílio tivessem dinheiro para comprar mais comida
contal = (tabsprenda.V2105.value_counts(sort=False, normalize=True)*100).round(decimals=2)
contal

```

```python
#Os moradores ficaram sem dinheiro para ter uma alimentação saudável e variada
contsem = (tabsprenda.V2107.value_counts(sort=False, normalize=True)*100).round(decimals=2)
contsem
```

```python
#Os moradores deste domicílio comeram apenas alguns alimentos que ainda tinham porque o dinheiro acabou
contpouco = (tabsprenda.V2109.value_counts(sort=False, normalize=True)*100).round(decimals=2)
contpouco
```

```python
# Gráfico de Setor
plot = contmora.plot(kind='bar',title='N° Moradores',figsize=(6, 6))
plt.ylabel('Freq.Relativa (em %)')
```

```python
plot = contprec.plot(kind='pie',title='Preocupação em acabar os alimentos',autopct='%.2f')
ax = plt.subplot(111)
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
plot = contal.plot(kind='pie',title='Os alimentos acabaram?',autopct='%.2f')
ax = plt.subplot(111)
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
plot = contsem.plot(kind='pie',title='Os moradores ficaram sem dinheiro para ter uma alimentação saudável e variada',autopct='%.2f')
ax = plt.subplot(111)
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
plot = contpouco.plot(kind='pie',title='Tiveram de comer apenas alguns alimentos que ainda tinham, porque o dinheiro acabou?',autopct='%.2f')
ax = plt.subplot(111)
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
